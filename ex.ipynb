```python
import pandas as pd
import numpy as np
from plotnine import ggplot, aes, geom_line

x = [1, 2, 3]
y = [[30, 4, 50], [300, 400, 500], [350, 450, 550]]

df = pd.DataFrame({
    "x": np.repeat(x, 3),
    "y": np.ravel(y),
    "g": list("abc" * 3)
})

print(df)

(ggplot(df, aes("x", "y", color="g"))
+ geom_line()
)
```
